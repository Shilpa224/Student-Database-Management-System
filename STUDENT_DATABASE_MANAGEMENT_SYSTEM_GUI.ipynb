```python
from tkinter import *
from tkinter import messagebox
from tkinter.ttk import *

# <--------------------------------------------- Login Window---------------------------------------------------->

def w1():
    window= Tk()
    window.title("USER LOGIN")
    window.configure(background="skyblue")
    lbl=Label(window,text="GLOBAL INSTITUTE OF MANAGEMENT & TECHNOLOGY",font=("Georgia bold",45),background="yellow",foreground="navy")
    lbl.place(x=90,y=50)
    lbl1=Label(window,text="STUDENT DATABASE",font=("Georgia",30),background="peachpuff",foreground="maroon")
    lbl1.place(x=750,y=200)
    window.geometry('1920x1080')
    
# <------User Name------->

    lbl2=Label(window,text="USERNAME",font=("Times New Roman Bold",20),background="lightpink",foreground="black")
    lbl2.place(x=650,y=370)
    ent2=Entry(window,width=30, font=("Times New Roman Bold",20))
    ent2.place(x=820,y=370)
    
# <------Password------->

    lbl3=Label(window,text="PASSWORD",font=("Times New Roman Bold",20),background="lightpink",foreground="black")
    lbl3.place(x=650,y=420)
    ent3=Entry(window,width=30, font=("Times New Roman Bold",20))
    ent3.place(x=820,y=420)
    
# <-----------Login Button Function---------->

    def clicked():
        if ent2.get()=="Admin" and ent3.get()=="2020":
            window.destroy()
            w2()
        else:
            messagebox.showerror('Error Messege','Error in Credential!!')
            
# <---------Login Button-------->

    btn=Button(window,text="Log In",command=clicked,width=30)
    btn.place(x=900,y=600)
    window.mainloop()
```

```python
from tkinter import *
import sqlite3 as sql

# <---------------------------------------------------Main Window---------------------------------------------------->

def w2():
    window = Tk()
    window.title("Registration")
    window.configure(background="salmon")
    window.geometry('1920x1080')
    
# <----------------------Window Heading--------------------->
    
    lableHeading=Label(window,text="STUDENT DETAILS",font=("Times New Roman Bold",30),background="cyan",foreground="black")
    lableHeading.place(x=800,y=50)
    
# <-------------------------------Student Name------------------------>  

    lableName=Label(window,text="NAME", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    lableName.place(x=200,y=300)
    
    student_name=Entry(window,width=40, font=("Times New Roman Bold",20))
    student_name.place(x=480,y=300)
    
# <-------------------Student Roll------------------>

    labelRoll=Label(window,text="ROLL", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelRoll.place(x=200,y=350)
    
    roll_no=Entry(window,width=40, font=("Times New Roman Bold",20))
    roll_no.place(x=480,y=350)
    
# <------------------Student Registration No--------------------->  

    labelREG_NO=Label(window,text="REG NO.", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelREG_NO.place(x=200,y=400)
    
    reg_no=Entry(window,width=40, font=("Times New Roman Bold",20))
    reg_no.place(x=480,y=400)
    
# <----------------Student Father Name---------------->

    labelFATHERS_NAME=Label(window,text="FATHER'S NAME", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelFATHERS_NAME.place(x=200,y=450)
    
    father_name=Entry(window,width=40, font=("Times New Roman Bold",20))
    father_name.place(x=480,y=450)
    
# <------------------Student Gender-------------------->

    labelGender=Label(window,text="GENDER", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelGender.place(x=200,y=500)

    gender=Combobox(window,width=39, font=("Times New Roman Bold",20))
    gender['values']=("select","MALE","FEMALE","OTHERS")
    gender.current(0)
    gender.place(x=480,y=500)
    
# <---------------------Student DOB----------------------->

    labelDOB=Label(window,text="D.O.B", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelDOB.place(x=200,y=550)
    
    dob=Entry(window,width=40, font=("Times New Roman Bold",20))
    dob.place(x=480,y=550)
    
# <--------------------Student ID------------------------>

    labelID=Label(window,text="STUDENT ID", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelID.place(x=200,y=600)
    
    student_id=Entry(window,width=40, font=("Times New Roman Bold",20))
    student_id.place(x=480,y=600)
    
# <-------------------Student Address-------------------->

    labelAddress=Label(window,text="ADDRESS", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelAddress.place(x=200,y=650)
    
    student_address=Entry(window,width=40, font=("Times New Roman Bold",20))
    student_address.place(x=480,y=650)
    
# <-------------------Student Mobile No.--------------------->

    labelMB_No=Label(window,text="MOBILE NO.", font=("Times New Roman Bold",20),background="paleturquoise",foreground="black")
    labelMB_No.place(x=200,y=700)
    
    ph_no=Entry(window,width=40, font=("Times New Roman Bold",20))
    ph_no.place(x=480,y=700)
    
#<---------------------Input--------------->

    label_input=Label(window,text="INPUT", font=("Times New Roman Bold",28),background="lightgray",foreground="black")
    label_input.place(x=600,y=200)
    

#<------------------Details-------------->

    label_details=Label(window,text="DETAILS", font=("Times New Roman Bold",28),background="lightgray",foreground="black")
    label_details.place(x=1480,y=200)
    
# <----------------Display Box--------------->

    t1=Text(window,width=90,height=30)
    t1.place(x=1100,y=300)
#<---------------------------------Database Part----------------------->    
    
    def connection():
        try:
            link = "C:/Users/user/Desktop/Python/Database/"
            conn = sql.connect(link +"Test1.db")
            print("Opened DataBase Successfully")
        except:
            print("Cannot connect to the database")
        return conn    


    def verifier():
        a=b=c=d=e=f=g=h=i=0
        if not student_name.get():
            t1.insert(END,"<>Student name is required<>\n")
            a=1
        if not roll_no.get():
            t1.insert(END,"<>Roll no is required<>\n")
            b=1
        if not reg_no.get():
            t1.insert(END,"<>Branch is required<>\n")
            c=1
        if not father_name.get():
            t1.insert(END,"<>Phone number is requrired<>\n")
            d=1
        if not gender.get():
            t1.insert(END,"<>Address is Required<>\n")
            e=1
        if not dob.get():
            t1.insert(END,"<>Address is Required<>\n")
            f=1
        if not student_id.get():
            t1.insert(END,"<>Address is Required<>\n")
            g=1
        if not student_address.get():
            t1.insert(END,"<>Address is Required<>\n")
            h=1
        if not ph_no.get():
            t1.insert(END,"<>Address is Required<>\n")
            i=1
        if a==1 or b==1 or c==1 or d==1 or e==1 or f==1 or g==1 or h==1 or i==1:
            return 1
        else:
            return 0

#<-----------------Add Button Functions---------------->
    def add_student():
                ret=verifier()
                if ret==0:
                    conn=connection()
                    cur=conn.cursor()
                    cur.execute("CREATE TABLE IF NOT EXISTS STUDENTS(NAME TEXT,ROLL_NO INTEGER,REG_NO TEXT,FATHER_NAME TEXT,GENDER TEXT,DOB TEXT,SID TEXT,ADDRESS TEXT,MOBILE TEXT)")
                    cur.execute("insert into STUDENTS values(?,?,?,?,?,?,?,?,?)",(student_name.get(),roll_no.get(),reg_no.get(),father_name.get(),gender.get(),dob.get(),student_id.get(),student_address.get(),ph_no.get()))
                    conn.commit()
                    conn.close()
                    t1.insert(END,"ADDED TO DATABASE SUCCESSFULLY\n")

# <----------------------Show All Button Function----------------->
    def view_student():
        conn=connection()
        cur=conn.cursor()
        cur.execute("select * from STUDENTS")
        data=cur.fetchall()
        conn.close()
        for i in data:
            t1.insert(END,str(i)+"\n")
#<---------------------Delete Button Function------------------->

    def delete_student():
        ret=verifier()
        if ret==0:
            conn=connection()
            cur=conn.cursor()
            cur.execute("DELETE FROM STUDENTS WHERE ROLL_NO=?",(int(roll_no.get()),))
            conn.commit()
            conn.close()
            t1.insert(END,"SUCCESSFULLY DELETED THE RECORD\n")

#<------------------Exit Button Function------------>
    def clse():
        window.destroy() 


#<------------------Buttons--------------------->

    btn1=Button(window,text="ADD",command=add_student,width=20)
    btn1.place(x=700,y=850)

    btn4=Button(window,text="SHOW ALL",command=view_student,width=20)
    btn4.place(x=500,y=850)


    btn5=Button(window,text="DELETE",command=delete_student,width=20)
    btn5.place(x=300,y=850)


    btn7=Button(window,text="EXIT",command=clse,width=20)
    btn7.place(x=900,y=850)
    window.mainloop()


```
